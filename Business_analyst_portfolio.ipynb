```python
#year in which it as highest sales
```

```python
# Step 1: Clean column names by stripping leading/trailing spaces
coffee_data.columns = coffee_data.columns.str.strip()

# Step 2: Clean and convert the 'Sales' column
coffee_data['Sales'] = coffee_data['Sales'].astype(str).str.replace('$', '').str.replace(',', '')
coffee_data['Sales'] = pd.to_numeric(coffee_data['Sales'])

# Step 3: Group by 'Year' and sum the 'Sales' for each year
annual_sales = coffee_data.groupby('Year')['Sales'].sum()

# Step 4: Find the year with the highest sales
year_highest_sales = annual_sales.idxmax()
highest_sales_amount = annual_sales.max()

# Display the result
print(f"The year with the highest sales is {year_highest_sales} with total sales of £{highest_sales_amount:.2f}M")
```

```python
#country performed highest in sales
```

```python
country_sales=coffee_data.groupby('Country')['Sales'].sum()
country_highest_sales=country_sales.idxmax()
country_highest_totalsales=country_sales.max()
print((country_highest_sales),"with sales of",(country_highest_totalsales))
```

```python
#the product generates highest revenue 
```

```python
product_revenue=coffee_data.groupby('Product')['Sales'].sum()
product_high_revenue=product_revenue.idxmax()
number_sales=product_revenue.max()
print("Product-",(product_high_revenue),"with sales of",(number_sales))

```

```python
#the product generates least revenue
```

```python
product_revenue = coffee_data.groupby('Product')['Sales'].sum()
product_low_revenue=product_revenue.idxmin()
product_low_sales=product_revenue.min()
print("Product-",(product_low_revenue),"with the sales of",(product_low_sales))
```

```python
#finding the segment with negative i.e lowest sales and postive i.e highest sales
```

```python
segment_revenue=coffee_data.groupby('Segment')['Sales'].sum()
segment_revenue_with_negative=segment_revenue.idxmin()
segment_revenue_with_postive=segment_revenue.idxmax()
print("Negative-",(segment_revenue_with_negative),",","Postive-",(segment_revenue_with_postive))
```

```python
import matplotlib.pyplot as plt

# Group by 'Country' and 'Product' to get the count of each product sold in each country
product_counts = coffee_data.groupby(['Country', 'Product']).size().unstack()

# Plot the data
product_counts.plot(kind='bar',figsize=(12, 6), colormap='viridis')
plt.xlabel('Product')
plt.ylabel('Count of Country Sold')
plt.title('Count of Each Product Sold by Product')
plt.legend(title='Product')
plt.show()
```
